## Merging Data Sets

```python
## Merging Customers and Orders Dataset
customer_level = pd.merge(orders, customer, how='left', on='customer_id')
```

```python
customer_level.shape
```

```python
avg_rating_orderwise = ratings.groupby('order_id').review_score.mean().reset_index()
```

```python
## Merging Customer Level and Avg. Ratings Orderwise Dataset
customer_level = customer_level.merge(avg_rating_orderwise, on = 'order_id')
```

```python
## Merging Order Items and Products Data Set
product_data = pd.merge(order_items, products, how='left', on='product_id')
```

```python
## Merging Seller with Product Data
product_level = product_data.merge(sellers, on='seller_id')
```

```python
product_level
```

```python
total_payments_orderwise = payments.groupby('order_id').payment_value.sum().reset_index()
```

```python
## Merging Customer Level with Orderwise Payments
customer_level = customer_level.merge(total_payments_orderwise, on='order_id')
```

```python
cust_order = pd.merge(customer_level, product_level[['order_id','product_category_name','seller_id','price','freight_value','order_item_id','product_id']], how='left', on='order_id')
```

```python
cust_order
```

```python
cust_order = cust_order.drop(columns=['order_approved_at','order_approved_at','order_delivered_carrier_date',
                                      'order_delivered_customer_date','order_estimated_delivery_date'])
```

```python
cust_order.info()
```

```python
cust_order.isna().sum()
```

### We can see that 4 columns have NaN values which we will be filling before we proceed with analysis

### For finding out the price and freight price based on payment value, we try to figure out average freight price over payment value which is sum of product price and freight price.

```python
freight_perc = (cust_order['freight_value'] / cust_order['payment_value'])*(100)
```

```python
freight_perc.plot(kind='hist')
```

```python
freight_perc.describe()
```

```python
avg_freight_perc = freight_perc.clip(lower = freight_perc.quantile(0.05), upper = freight_perc.quantile(0.95)).mean()
avg_freight_perc
```

### After Doing Outlier Treatment we can round the freight price to 18% of payment value so the product price will be 82%  of the payment value. Now lets fill the NaN values with approriate prices.

```python
cust_order.price = cust_order.price.fillna(round(0.82*cust_order.payment_value,2))
```

```python
cust_order.freight_value = cust_order.freight_value.fillna(round(0.18*cust_order.payment_value,2))
```

```python
cust_order.info()
```

### Now lets fill NaN values in order_item_id. It signifies how many times the order was repeated and it increments by 1 after it is repeated.

```python
cust_order.order_item_id.plot(kind='hist')
```

```python
## Since most of the order were unique and not repeated more than once, we can fill the na values with the mode
cust_order.order_item_id = cust_order.order_item_id.fillna(cust_order.order_item_id.mode()[0])
```

```python
cust_order.info()
```

### Now, since seller id and product_category_name are not known for some rows we can mark them as 'unknown'

```python
for i in ['product_category_name', 'seller_id']:
    cust_order[i] = cust_order[i].fillna('Unknown')
```

```python
cust_order.info()
```

### Converting order_purchase_timestamp to datetime

```python
cust_order.order_purchase_timestamp = pd.to_datetime(cust_order.order_purchase_timestamp)
```

### Adding new Date Columns for better analysis

```python
cust_order['year']=cust_order.order_purchase_timestamp.dt.year
```

```python
cust_order['month']=cust_order.order_purchase_timestamp.dt.month_name()
```

```python
cust_order['week']=cust_order.order_purchase_timestamp.dt.weekofyear
```

```python
cust_order['time']=cust_order.order_purchase_timestamp.dt.time
```

```python
cust_order['day']=cust_order.order_purchase_timestamp.dt.date
```

```python
cust_order['month_year']=cust_order.order_purchase_timestamp.dt.date.astype(str).str[0:7]
```

```python
## Since customers can avail discount, we add one more column for how much they have to pay before applying discount
cust_order['price_to_pay']=cust_order['price']+cust_order['freight_value']
```

## 2. Performing Customers/sellers Segmentation
### a. Divide the customers into groups based on the revenue generated

```python
a2=cust_order.groupby('customer_id').price_to_pay.sum().reset_index()
a2['Group_number']=pd.qcut(a2['price_to_pay'],3,labels=['Low Spenders','Moderate Spenders','High Spenders'])
a2[['customer_id','Group_number']]
```

#### The Customers are segmented in 3 groups, Low Spenders, Moderate Spenders and High Spenders. This is enable the business marketing team to device marketing campaigns accordingly.

### b. Divide the sellers into groups based on the revenue generated

```python
b2=cust_order.groupby('seller_id').price_to_pay.sum().reset_index()
b2['Group_number']=pd.qcut(b2['price_to_pay'],3,labels=['Gold Category','Silver Category','Bronze Category'])
b2[['seller_id','Group_number']]
```

#### The Sellers are segmented in 4 groups Gold Category -> Top Sellers with most Revenue Generated, Silver Category -> Sellers with moderate Revenue Generated and Bronze Sellers -> Sellers with Least Revenue Generated

## 3. Cross-Selling (Which products are selling together)

```python
a3 = pd.crosstab(cust_order.order_id, cust_order.product_category_name)

for i in a3.columns:
    a3[i] = a3[i].apply(lambda x: True if x>0 else False)
```

```python
from mlxtend.frequent_patterns import fpmax

fpmax(a3,min_support=0.00001,use_colnames=True).tail(10).itemsets.iloc[::-1].reset_index(drop=True)
```

### The Above combination of 2 product categories are most commonly sold together
